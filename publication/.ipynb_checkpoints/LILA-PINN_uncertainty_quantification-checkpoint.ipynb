```python
import sys
sys.path.append('..')
from _utils import *
from _utils.methods import *
u_path = '../_utils/'
```

```python
n_models = 100
s=3
```

```python
#%% Leak 1 DMA C

# leak specific parameters
name = 'l1'
leak_start = '2019-02-10 13:05:00'

train_start = '2019-01-01'
train_end   = '2019-01-14'
test_start  = '2019-01-01'
test_end    = '2019-03-31'

# evaluation
df_plot_TTD_l1 = pd.DataFrame()
df_plot_metrics_l1 = pd.DataFrame(columns=['Precision','Recall','F1_score'])
TP,FN,FP = [0,0,0]

with open(u_path+'list_det_{}_s{}.pkl'.format(name,s), 'rb') as f:
    list_det_QNET = pickle.load(f)

for i in range(n_models):
    if list_det_QNET[i].empty or (list_det_QNET[i].dropna().empty) or ((list_det_QNET[i].iloc[0]-pd.to_datetime(leak_start)).total_seconds()>3600*24*2):
        FN+=1
    elif list_det_QNET[i].iloc[0] < pd.to_datetime(leak_start):
        FP+=1
    else:
        df_plot_TTD_l1.loc[i,s] = (list_det_QNET[i].iloc[0] - pd.to_datetime(leak_start)).total_seconds()/3600
        TP+=1
df_plot_metrics_l1.loc[s,'TP'] = TP
df_plot_metrics_l1.loc[s,'FP'] = FP
df_plot_metrics_l1.loc[s,'FN'] = FN
df_plot_metrics_l1.loc[s,'Precision'] = TP/(TP+FP)
df_plot_metrics_l1.loc[s,'Recall'] = TP/(TP+FN)
df_plot_metrics_l1.loc[s,'F1_score'] = 2*TP/(2*TP+FN+FP)
```

```python
#%% Leak 2 DMA C

# leak specific parameters
name = 'l2'
leak_start = '2019-05-30 21:55:00'

train_start = '2019-04-01'
train_end   = '2019-04-14'
test_start  = '2019-04-01'
test_end    = '2019-10-31'

# evaluation
df_plot_TTD_l2 = pd.DataFrame()
df_plot_metrics_l2 = pd.DataFrame(columns=['Precision','Recall','F1_score'])
TP,FN,FP = [0,0,0]

with open(u_path+'list_det_{}_s{}.pkl'.format(name,s), 'rb') as f:
    list_det_QNET = pickle.load(f)

for i in range(n_models):
    if list_det_QNET[i].empty or (list_det_QNET[i].dropna().empty):
        # or ((list_det_QNET[i].iloc[0]-pd.to_datetime(leak_start)).total_seconds()>3600*24*100):
        FN+=1
    elif list_det_QNET[i].iloc[0] < pd.to_datetime(leak_start):
        FP+=1
    else:
        df_plot_TTD_l2.loc[i,s] = (list_det_QNET[i].iloc[0] - pd.to_datetime(leak_start)).total_seconds()/3600/24
        TP+=1
df_plot_metrics_l2.loc[s,'TP'] = TP
df_plot_metrics_l2.loc[s,'FP'] = FP
df_plot_metrics_l2.loc[s,'FN'] = FN
df_plot_metrics_l2.loc[s,'Precision'] = TP/(TP+FP)
df_plot_metrics_l2.loc[s,'Recall'] = TP/(TP+FN)
df_plot_metrics_l2.loc[s,'F1_score'] = 2*TP/(2*TP+FN+FP)
```

### Figure 6

```python
f,axs=plt.subplots(2,figsize=(15,10))
for i,(ax,df_plot,df_metrics) in enumerate(zip(axs,
                                 [df_plot_TTD_l1,df_plot_TTD_l2],
                                 [df_plot_metrics_l1,df_plot_metrics_l2]
                                )):
    data = df_plot.loc[:,s].dropna()
    items = ax.violinplot(data,positions=[s],showextrema=False, showmedians=False, vert=False)
    for pc in items['bodies']:
        pc.set_facecolor(mpl.colormaps.get_cmap('Pastel1')(0.3))
        pc.set_edgecolor('black')
        pc.set_alpha(1)
    quartile1, medians, quartile3 = np.percentile(data.values, [25, 50, 75])
    whiskers_min, whiskers_max = data.min(), data.max()

    ax.scatter(medians,s, marker='o', color='white', s=80, zorder=3)
    ax.hlines(s, quartile1, quartile3, color='k', linestyle='-', lw=15)
    ax.hlines(s, whiskers_min, whiskers_max, color='k', linestyle='-', lw=1)
    
    ax.set_yticks([3])
    if i==0:
        ax.text(20.02,3.23,'({})'.format(string.ascii_lowercase[i]))
        ax.set_xlabel('$TTD$ ($h$)')
        ax.set_ylabel('Abrupt leak')
        ax.set_yticklabels([])
    else:
        ax.text(2.92,3.23,'({})'.format(string.ascii_lowercase[i]))
        ax.set_xlabel('$TTD$ ($d$)')
        ax.set_ylabel('Incipient leak')
        ax.set_yticklabels([])
f.tight_layout()
```
